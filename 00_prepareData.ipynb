```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = np.genfromtxt('852.336nm_21042023_16.13.csv', delimiter=',')
```

```python
%matplotlib notebook
plt.close()
plt.plot(data[:,0])
# plt.plot((data[:,2]-np.mean(data[:,2]))/1000)
plt.show()
```

```python
min_index = np.where(data[:,0] == np.min(data[:,0]))
max_index = np.where(data[:,0] == np.max(data[:,0]))
```

```python
print(f'min_index length {np.shape(min_index)[1]}\nmax_index length {np.shape(max_index)[1]}')
```

```python
min_index, max_index
```

```python
for i in range(0, np.shape(max_index)[1]-1):
    ramp_0 = data[min_index[0][i]:max_index[0][i+1],0]
    pd1_0 = data[min_index[0][i]:max_index[0][i+1],1]
    pd2_0 = data[min_index[0][i]:max_index[0][i+1],2]
    a = np.hstack((np.reshape(np.asarray(ramp_0),(len(ramp_0), 1)), np.reshape(np.asarray(pd1_0),(len(pd1_0), 1)), np.reshape(np.asarray(pd2_0),(len(pd2_0), 1))))
    np.savetxt("./852.336nm_21042023_16.16_8.2nF filter/data_"+str(i)+".csv", a, delimiter=",")
#     print(min_index[0][i], max_index[0][i+1])
```

```python
for i in range(1, np.shape(max_index)[1]-1):
    ramp_0 = data[min_index[0][i]:max_index[0][i-1],0]
    pd1_0 = data[min_index[0][i]:max_index[0][i-1],1]
    pd2_0 = data[min_index[0][i]:max_index[0][i-1],2]
    a = np.hstack((np.reshape(np.asarray(ramp_0),(len(ramp_0), 1)), np.reshape(np.asarray(pd1_0),(len(pd1_0), 1)), np.reshape(np.asarray(pd2_0),(len(pd2_0), 1))))
    np.savetxt("./852.336nm_21042023_16.16_8.2nF filter/data_"+str(i)+".csv", a, delimiter=",")
#     print(min_index[0][i], max_index[0][i+1])
```
